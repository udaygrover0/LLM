```python
# 2.1 Where is our data file:
pathToFolder = "/gdrive/MyDrive/deeplearning/deeplearning"
```

```python
# 2.2 Full file path:
path = Path(pathToFolder) / "red_wine.csv"
```

```python
# 3.0 Read data
red_wine = pd.read_csv(path)
red_wine.head()
```

```python
# prompt: Using dataframe red_wine: For dataset red wine please let me  know if there is any correlation between residual sugar and alcohol

import pandas as pd

# Calculate the correlation coefficient between residual sugar and alcohol
correlation = red_wine['residual sugar'].corr(red_wine['alcohol'])

# Print the correlation coefficient
print(f"The correlation coefficient between residual sugar and alcohol is {correlation:.3f}")
```

```python
# prompt: for same dataset please tell me if there is any corelation between Ph and alcohol

correlation = red_wine['alcohol'].corr(red_wine['pH'])
print(f"The correlation coefficient between alcohol and pH is {correlation:.3f}")
```

```python
# 3.1
red_wine.shape   # (1599, 12)
red_wine.dtypes
```

```python
# 3.2 Get target and predictors:
y = red_wine.pop("quality")
X = red_wine
```

```python
# 4.0 Build model:

model = tf.keras.Sequential()
```

```python
# 4.1
model.add(layers.Input( shape = (11,) ) )
```

```python
# 4.2
model.add(layers.Dense(1))
```

image.png

```python
# 4.2.1
weights, bias = model.weights
```

```python
# 4.2.2
weights
```

```python
# 4.2.3
bias
```

```python
#4.3 Compile
model.compile( loss = "mae")
```

```python
# 4.4 Train
model.fit(X,y, epochs = 100 )
```

```python
# 5.0 Predict
pred = model.predict(X)
```

```python
# 5.1 Round up decimal places
pred = np.round(pred)
```

```python
# 5.1.1
y.shape
```

```python
# 5.2 Compare:
np.sum(pred == y.values.reshape(1599,1))/1599
```
